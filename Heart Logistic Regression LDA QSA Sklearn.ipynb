```python
import pandas as pd
from pandas import read_csv
import numpy as np
import sklearn.linear_model as lm
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
```

```python
filename = 'heart.csv'
df = read_csv(filename)
```

```python
df
```

```python
heart_data = df.values
```

```python
X = heart_data[:, 0:12]
Y = heart_data[:, 13]
scaler = MinMaxScaler(feature_range=(0, 1))
rescaled_X = scaler.fit_transform(X)
```

```python
my_test_size = 0.2
seed = 2 
X_train, X_test, Y_train, Y_test = train_test_split(rescaled_X, Y, test_size = my_test_size , random_state = seed)
```

```python
logreg = lm.LogisticRegression(C = 1e8, solver = 'lbfgs')
```

```python
logreg.fit(X_train, Y_train)
Y_pred_logreg = logreg.predict(X_test)
errors = Y_pred_logreg != Y_test
print("Nb errors=%i, error rate=%.2f" % (errors.sum(), errors.sum() / len(Y_pred_logreg)))
print(logreg.coef_)
```

```python
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis as LDA
lda = LDA()
lda.fit(X_train, Y_train)
Y_pred_lda = lda.predict(X_test)
errors = Y_pred_lda != Y_test
print("Nb errors=%i, error rate=%.2f" % (errors.sum(), errors.sum() / len(Y_pred_lda)))
print(lda.coef_)
```

```python
from sklearn.discriminant_analysis import QuadraticDiscriminantAnalysis as QDA
qda = QDA()
qda.fit(X_train, Y_train)
Y_pred_qda = qda.predict(X_test)
errors = Y_pred_qda != Y_test
print("Nb errors=%i, error rate=%.2f" % (errors.sum(), errors.sum() / len(Y_pred_qda)))
```
